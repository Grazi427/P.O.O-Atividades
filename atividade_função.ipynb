```python
global estoque_bebida
global estoque_comida

estoque_comida={'pão de queijo':15,'sardinha':8,'macarrão':20,'sanduíche':33,'arroz':40,'pastel':24,'banana':10,'maracuja':7,'frango':10,'feijão':26,'pizza':30}
estoque_bebida={'coca-cola':40,'suco de maracuja':27,'suco de laranja':30,'achocolatado':33,'energetico':18,'água':28,'milkshake':20,'café':30,'sprite':24,'café com leite':6,'água com gás':8}

def menu():
    print('================= MENU DE GERANCIAMENTO ================\n')
    print('1 - Mostrar estoque '
        '\n2 - Adicionar produto '
        '\n3 - Remover produto '
        '\n4 - Consultar produto '
        '\n5 - Salvar relatório de estoque '
        '\n6 - Repor estoque automático '
        '\n7 - Sair do sistema ')
    print('==========================================================\n')
    esc=input('Digite o número que deseja acessar:  ')
    if esc=='1':
        mostrar_estoque()
    elif esc=='2':
        adicionar_produto()
    elif esc=='3':
        remover_produto()
    elif esc=='4':
        consultar_produto()
    elif esc=='5':
        salvar_estoque()
    elif esc=='6':
        repor_automatico()
    elif esc=='7':
        print('Sistema encerrado!')
        exit()
    else:
        print('Opção inválida')
        menu()


def mostrar_estoque():
    print('===================== ESTOQUE DE PRODUTOS ====================')
    for nome, quantidade in estoque_comida.items():
        print(f'{nome}: {quantidade}')
    for nome, quantidade in estoque_bebida.items():
        print(f'{nome}: {quantidade}')
    return menu()

def adicionar_produto():
    escolha=input('\nVocê deseja:'
          '\n1 - adicionar novo produto '
          '\n2 - adicionar estoque\n')

    tipo = input('\nO produto é:'
                 '\n 1 - Bebida'
                 '\n 2 - Comida\n')
    if escolha == '1':
        nome=input('Digite o nome do produto:  ')
        quantidade=int(input('Digite a quantidade do produto:  '))
        if tipo=='1':
          estoque_bebida[nome]=quantidade
        elif tipo=='2':
            estoque_comida[nome]=quantidade
        else:
            print('\nTipo de produto inválido')
    elif escolha=='2':
        nome = input('Digite o nome do produto que deseja atualizar:  ')
        quantidade = int(input('Digite a quantidade que deseja adiconar:  '))
        if tipo=='1':
          estoque_bebida[nome]+=quantidade
        elif tipo=='2':
            estoque_comida[nome]+=quantidade
        else:
            print('\nTipo de produto inválido')
    return menu()


def remover_produto():
  tipo = input('\nO produto é:'
                 '\n 1 - Bebida'
                 '\n 2 - Comida')
  nome = input('\nDigite o nome do produto que deseja atualizar:  ')
  quantidade = int(input('Digite a quantidade que deseja remover:  '))
  if tipo == '1':
      estoque_bebida[nome]-=quantidade
  elif tipo == '2':
      estoque_comida[nome]-=quantidade
  else:
      print('\nTipo de produto inválido')
  return menu()

def consultar_produto():
    nome = input('\nQual produto deseja verificar:  ')
    if nome in estoque_comida.items():
        print(estoque_comida[nome])
    elif nome in estoque_bebida.items():
        print(estoque_bebida[nome])
    else:
        print('\nProduto não encontrado')
    return menu()

def salvar_estoque():
  with open('estoque.txt', 'w'):
    for nome, quantidade in estoque_comida.items():
      with open('estoque.txt', 'a') as f:
        f.write(f'{nome}: {quantidade}\n')

    for nome, quantidade in estoque_bebida.items():
      with open('estoque.txt', 'a') as f:
        f.write(f'{nome}: {quantidade}\n')
    return menu()


def repor_automatico():
  for nome, quantidade in estoque_comida.items():
    if quantidade<=3:
      estoque_comida[nome]+=5

  for nome, quantidade in estoque_bebida.items():
    if quantidade<=3:
      estoque_bebida[nome]+=5
  return menu()


menu()
```
